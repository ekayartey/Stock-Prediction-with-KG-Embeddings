```python
from twitterscraper import query_tweets
import datetime as dt
import pandas as pd
import os
```

```python
begin_date=dt.date(2019,1,1)
end_date=dt.date(2019,12,31)
limit=100000
lang='english'
```

```python
tweets=query_tweets("#MSFT OR #$MSFT OR $MSFT OR MSFT OR Microsoft from:wsj OR from:reuters OR from:business OR from:cnbc OR from:RANsquawk OR from:wsjmarkets",
                    limit=limit,lang=lang)
```

```python
StockNewsTweets =pd.DataFrame(t.__dict__ for t in tweets)
df = StockNewsTweets.copy()
```

```python
df = df[['timestamp', 'text', 'username']]
df.set_index(pd.to_datetime(df['timestamp']), inplace=True)
df.drop(columns='timestamp', inplace=True)

df
```

```python
df.drop_duplicates(subset=['text'], keep=False, inplace=True)
df
```

```python
# Check if stock news data exit
if not os.path.exists('./msft_news_twitter.csv'):

# make directory and store stock news    
    #os.makedirs('stockdata')
    df.to_csv('./msft_news_twitter.csv')
else:
    pass
```
